```python
import pandas as pd
import pydotplus
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from IPython.display import Image  
```

```python
read_diabetics_data = pd.read_csv('pima-indians-diabetes.csv')
```

```python
read_diabetics_data.columns =['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class_names']
```

```python
read_diabetics_data.head(5)
```

```python
# converting the overall data into features and target
X = read_diabetics_data.iloc[:, 0:read_diabetics_data.shape[1]-1]
y = read_diabetics_data.iloc[:,read_diabetics_data.shape[1]-1]
```

```python
"""
Concept behind Decision tree

ROOT NODE: main Node from where one get other nodes such as leaf and branch nodes appears
Branch NODE: it is the node that is derived from Root Node
LEAF NODE: It is the node which cannot be further splitted
These Nodes contains the features and root node is chosen based on information gain

Information Gain = Overall Entropy - Independent entropy of all input variables

The variable whose entropy is minimum is chosen as the Root node because in that
case the information gain is maximum
"""
```

```python
tree = DecisionTreeClassifier(criterion = 'entropy').fit(X, y)
```
